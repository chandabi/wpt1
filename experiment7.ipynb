```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
cars_df=pd.read_csv('mtcars (4).csv')
print(cars_df)
```

```python
trans=cars_df['am'].value_counts()
print(trans)
trans.plot.bar()
plt.xlabel("trasmission")
plt.title('BarplottocheckthefrequencydistributionofTransmissiontype')
plt.show()
```

```python
trans=cars_df['am'].value_counts()
print(trans)
trans.plot.bar()
plt.xlabel("transmission")
plt.title('BarplottocheckthefrequencydistributionofTransmissiontype')
plt.show()
```

```python
a=plt.scatter(cars_df['mpg'], cars_df['wt'], marker='*')
print(a)
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.title('Relationship between the weight of the cars and MPG')
plt.show()
```

```python
cars_df['mpg'].plot(kind='box')
plt.xlabel('MPG')
plt.title('Box and whisker plot of MPG')
plt.show()
```

```python
minvalue=np.min(cars_df['mpg'])
q1=np.percentile(cars_df['mpg'],25)
median=np.median(cars_df['mpg'])
q3=np.percentile(cars_df['mpg'],75)
maxvalue=np.max(cars_df['mpg'])
print('minimum value:',minvalue)
print('Q1:',q1)
print('median:',median)
print('Q3:',q3)
print('maximum value:',maxvalue)
```
