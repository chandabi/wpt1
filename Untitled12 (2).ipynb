```python
import pandas as pd
import numpy as np
data={
      'First':['Aryan','Rachan','Riya','Yash','Siddini'],
      'Lastname':['Singh','Agarwal','Shah','Batia','Khan'],
      'Type':['Full_time Employee','Inter','Full_time Employee','Part_time Employee','Full_time Employee'],
      'Department':['Administration','Technical','Administration','Technical','Technical'],
      'Yoe':[2,3,5,7,6],
      'Salary':[20000,5000,10000,10000,20000]
     }
df=pd.DataFrame(data)
print(df)
table=pd.pivot_table(df,index=['Type',],
values='Salary',aggfunc='mean')
print(table)
table=pd.pivot_table(df,index=['Type'],
values='Salary',aggfunc=['count'])
print(table)
table=pd.pivot_table(df,index=['Type'],
values='Salary',aggfunc=['std'])
print(table)
```
