```python
import pandas as pd
import pandas as np
data={
    'First':['aryan','rachan','riya','yash','sidhini'],
    'LastName':['singh','agarwal','shah','batia','khan'],
    'type':['Full_time Employee','Inter','Full_time Employee','part_time Employee','Full_time Employee'],
     'Department':['Administration','Technical','Administration','Technical','technical'],
                'yoe':[2,3,5,6,7],
                'salary':[20000,50000,10000,10000,200000]
}
df=pd.DataFrame(data)
print(df)
table=pd.pivot_table(df,index=['type','Department'],values='salary',aggfunc='mean')
print(table)
table1=pd.pivot_table(df,index=['type'],values='salary',aggfunc=['mean','sum'])
print(table)
table2=pd.pivot_table(df,index=['type'],values='salary',aggfunc=['count'])
print(table2)
table3=pd.pivot_table(df,index=['type'],values='salary',aggfunc=['std'])
print(table3)


```
